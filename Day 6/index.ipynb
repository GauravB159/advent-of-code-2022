```python
# Read input from file
with open('input.txt', 'r') as f:
    input = f.read()
input = list(filter(lambda x: x, [(input[i:i+4], i + 4) if len(set(input[i:i+4])) == 4 else 0 for i,x in enumerate(input)]))[0][1]
print(input)
```
